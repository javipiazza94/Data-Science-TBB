```python
population = population.loc[population['Countries'].notna()]
```

```python
population['Countries'] = population['Countries'].map(countries)
```

```python
population['Demographic indicator'] = population['Demographic indicator'].map(indicators_dict)
```

```python
population
```

```python
unique_values = population['Demographic indicator'].unique()
subtables = {}
for value in unique_values:
    subtable = population[population['Demographic indicator'] == value]
    subtables[value] = subtable
```

```python
for value, subtable in subtables.items():
    filename = f'./data/{value}.csv'
    subtable.to_csv(filename, index=False)
    print(f"Guardado: {filename}")
```
