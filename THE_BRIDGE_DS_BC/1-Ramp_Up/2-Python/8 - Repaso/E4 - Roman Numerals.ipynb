# Tu código

````python
class RomanNumerals:
    miles = ["", "M", "MM", "MMM", "IV̅", "V̅", "V̅M", "V̅MM", "V̅MMM", "IX̅"]
    centenas = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
    decenas = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
    unidades = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]

    miles_decimales = {"": 0, "M": 1000, "MM": 2000, "MMM": 3000, "IV̅": 4000, "V̅": 5000, "V̅M": 6000, "V̅MM": 7000, "V̅MMM": 8000, "IX̅": 9000}
    centenas_decimales = {"": 0, "C": 100, "CC": 200, "CCC": 300, "CD": 400, "D": 500, "DC": 600, "DCC": 700, "DCCC": 800, "CM": 900}
    decenas_decimales = {"": 0, "X": 10, "XX": 20, "XXX": 30, "XL": 40, "L": 50, "LX": 60, "LXX": 70, "LXXX": 80, "XC": 90}
    unidades_decimales = {"": 0, "I": 1, "II": 2, "III": 3, "IV": 4, "V": 5, "VI": 6, "VII": 7, "VIII": 8, "IX": 9}

    def __init__(self):
        pass

    def decimal_to_roman(self, numero: int):
        miles = self.miles[numero // 1000]
        centenas = self.centenas[(numero % 1000) // 100]
        decenas = self.decenas[(numero % 100) // 10]
        unidades = self.unidades[numero % 10]
        resultado = miles + centenas + decenas + unidades
        return resultado

    def roman_to_decimal(self, letras: str):
        resultado = 0
        i = 0

        while i < len(letras):
            found = False  # Un indicador para verificar si hemos encontrado una secuencia válida.

            # Comenzamos buscando secuencias de 3 caracteres y luego disminuimos a 2 y 1 caracter.
            for j in range(3, 0, -1):
                if letras[i:i + j] in self.miles_decimales:
                    resultado += self.miles_decimales[letras[i:i + j]]  # Sumamos el valor correspondiente.
                    i += j  # Avanzamos en el índice i en función del número de caracteres encontrados.
                    found = True  # Indicamos que encontramos una secuencia válida.
                    break

            # Si no encontramos una secuencia válida en miles, continuamos buscando en centenas, decenas y unidades de manera similar.
            if not found:
                for j in range(2, 0, -1):
                    if letras[i:i + j] in self.centenas_decimales:
                        resultado += self.centenas_decimales[letras[i:i + j]]
                        i += j
                        found = True
                        break

            if not found:
                for j in range(2, 0, -1):
                    if letras[i:i + j] in self.decenas_decimales:
                        resultado += self.decenas_decimales[letras[i:i + j]]
                        i += j
                        found = True
                        break

            if not found:
                for j in range(2, 0, -1):
                    if letras[i:i + j] in self.unidades_decimales:
                        resultado += self.unidades_decimales[letras[i:i + j]]
                        i += j
                        found = True
                        break

            if not found:
                raise ValueError("Número romano no válido")  # Si no encontramos una secuencia válida, lanzamos una excepción.

        return resultado    
````

````python
x = RomanNumerals()
y = x.decimal_to_roman(8769)
z = x.roman_to_decimal("MMMDCCLXVII")

print(y)  # Debería imprimir "V̅III̅DCCLXIX"
print(z)  # Debería imprimir 3767
````
