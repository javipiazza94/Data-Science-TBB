6. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gr치ficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisi칩n en el eje X que desees.  

CONSEJO --> Pintar por capas

```python
# Valores del eje X entre 0 y 5 con una precisi칩n de 0.1
x_vals = np.arange(0, 5.1, 0.1)

# Ecuaci칩n de la recta con pendiente positiva de 3 que pasa por (0,0)
y_vals_positiva = 3 * x_vals

# Ecuaci칩n de la recta con pendiente negativa de 3 que pasa por (0, -5)
y_vals_negativa = -3 * x_vals - 5

# Crear la figura y los ejes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Graficar la recta con pendiente positiva en el primer conjunto de ejes
ax1.plot(x_vals, y_vals_positiva, label='Recta con pendiente positiva de 3')
ax1.set_title('Recta con pendiente positiva')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')
ax1.legend()

# Graficar la recta con pendiente negativa en el segundo conjunto de ejes
ax2.plot(x_vals, y_vals_negativa, label='Recta con pendiente negativa de -3')
ax2.set_title('Recta con pendiente negativa')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')
ax2.legend()

# Ajustar el dise침o para evitar solapamientos
plt.tight_layout()

```

8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es m치s sencillo dejarlos dentro de la figura.

```python
# Eje X
personas = np.array(['Ana', 'Luis', 'Pedro', 'Sofia', 'Carmen'])

#Eje Y
alturas = np.array([160, 180, 175, 190, 170])

#Etiquetas
plt.xlabel('Nombres')
plt.ylabel('Altura en centimetros')
plt.title('Altura por persona')

#Grafico
grafico4 = plt.bar(personas, alturas, color='green', alpha=0.7)
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
barplot = plt.bar(personas, alturas, color='green')

for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval+1, int(yval), ha='center', va='bottom')

# plt.bar(personas, alturas, color='green')
plt.xlabel("Personas")
plt.ylabel("Alturas")
plt.title("Alturas grupo amigos")
            
```
