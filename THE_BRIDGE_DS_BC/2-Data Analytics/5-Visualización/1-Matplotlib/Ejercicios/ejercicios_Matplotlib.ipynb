6. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
# Valores del eje X entre 0 y 5 con una precisión de 0.1
x_vals = np.arange(0, 5.1, 0.1)

# Ecuación de la recta con pendiente positiva de 3 que pasa por (0,0)
y_vals_positiva = 3 * x_vals

# Ecuación de la recta con pendiente negativa de 3 que pasa por (0, -5)
y_vals_negativa = -3 * x_vals - 5

# Crear la figura y los ejes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Graficar la recta con pendiente positiva en el primer conjunto de ejes
ax1.plot(x_vals, y_vals_positiva, label='Recta con pendiente positiva de 3')
ax1.set_title('Recta con pendiente positiva')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')
ax1.legend()

# Graficar la recta con pendiente negativa en el segundo conjunto de ejes
ax2.plot(x_vals, y_vals_negativa, label='Recta con pendiente negativa de -3')
ax2.set_title('Recta con pendiente negativa')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')
ax2.legend()

# Ajustar el diseño para evitar solapamientos
plt.tight_layout()

```

9. Basándote en el gráfico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los métodos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
# Eje X
personas = np.array(['Ana', 'Luis', 'Pedro', 'Sofia', 'Carmen'])

#Eje Y
alturas = np.array([160, 180, 175, 190, 170])

#Etiquetas
plt.xlabel('Nombres')
plt.ylabel('Altura en centimetros')
plt.title('Altura por persona')

#Grafico
grafico5 = plt.bar(personas, alturas, color='green', alpha=0.7)

#Se puede recorrer cada barra en un bucle al haberle asignado una variable al grafico
for bar in grafico5:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', 
             ha='center', va='bottom')
```

12. Prueba a hacer el gráfico con las barras en horizontal (no es necesario poner el texto al final de la barra).

```python
# Crear el gráfico de barras horizontal con un tamaño de figura de 3x5 pulgadas
fig, ax = plt.subplots(figsize=(5, 5))

# Crear el gráfico de barras horizontal
grafico8 = ax.barh(x, y, color='green', alpha=0.7)

# Ajustar los límites del eje x
ax.set_xlim(0, max(y) + 10)

# Agregar etiquetas de alturas encima de cada barra con tamaño de fuente 3.5
for bar in grafico8:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

# Etiquetas
ax.set_xlabel('Altura en centímetros')
ax.set_ylabel('Nombres')
ax.set_title('Altura por persona')
```

13. Prueba a invertir los ejes con ax.invert_axis()

```python
# Crear el gráfico de barras horizontal con un tamaño de figura de 3x5 pulgadas
fig, ax = plt.subplots(figsize=(5, 5))

# Crear el gráfico de barras horizontal
grafico9 = ax.barh(x, y, color='green', alpha=0.7)

# Ajustar los límites del eje x
ax.set_xlim(0, max(y) + 10)

# Agregar etiquetas de alturas encima de cada barra con tamaño de fuente 3.5
for bar in grafico9:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

# Etiquetas
ax.set_xlabel('Altura en centímetros')
ax.set_ylabel('Nombres')
ax.set_title('Altura por persona')

#Invertimos ejes --> Va al final y no al principio
ax.invert_xaxis()
```

14. En el último gráfico, cambia el estilo a 'dark_background'

```python

# Crear el gráfico de barras horizontal con un tamaño de figura de 3x5 pulgadas
fig, ax = plt.subplots(figsize=(5, 5))

# Crear el gráfico de barras horizontal
grafico10 = ax.barh(x, y, color='green', alpha=0.7)

# Ajustar los límites del eje x
ax.set_xlim(0, max(y) + 10)

# Aplicar estilo "darkgrid" de seaborn
plt.style.use('dark_background')

# Agregar etiquetas de alturas encima de cada barra con tamaño de fuente 3.5
for bar in grafico10:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', 
            ha='center', va='bottom')

# Etiquetas
ax.set_xlabel('Altura en centímetros')
ax.set_ylabel('Nombres')
ax.set_title('Altura por persona')

#Invertimos ejes --> Va al final y no al principio
ax.invert_xaxis()
```
