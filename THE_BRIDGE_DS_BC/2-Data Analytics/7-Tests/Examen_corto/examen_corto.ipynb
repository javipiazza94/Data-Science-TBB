# Ejercicio de CAMPEONES

Carga el archivo `detalles.csv` en un dataframe de pandas.

Dentro de la columna `number_of_observations_food` se encuentra un listado de elementos. Esto se puede entender como que dentro tiene un Dataframe. 

Tu objetivo es hacer que el contenido de este campo se conviertan en columnas del dataframe. 

Ejemplo: La primera fila tiene dentro de esa columna los valores `'bread: 2318, rice: 2701, wheat: 2682'`. Tu dataframe ahora tendrá 3 columnas más, y rellenar los valores correspondientes de cada fila en cada columna.

```python
df = pd.read_csv('./data/detalles.csv')
df.head(5)
```

```python
df['number_of_observations_food']
```

```python
def parser(diccionario_df):
    diccionario_parseado = {}
    elementos = diccionario_df.split(',')
    for elemento in elementos:
        elemento = elementos.split(':')
        diccionario_parseado[elemento[0]]= elemento[1]
    return diccionario_parseado

```

```python

# Crear un DataFrame vacío para almacenar los resultados
resultados = pd.DataFrame()

for fila in range(0, 25):
    diccionario = df['number_of_observations_food'][fila]
    diccionario_parseado = {}
    elementos = diccionario.split(',')
    for elemento in elementos:
        clave, valor = elemento.split(':')
        diccionario_parseado[clave.strip()] = int(valor.strip())
    
    # Crear un DataFrame temporal con el diccionario parseado
    ingredientes = pd.DataFrame(data=[diccionario_parseado])
    
    # Concatenar el DataFrame temporal al DataFrame de resultados
    resultados = pd.concat([resultados, ingredientes], ignore_index=True)

# Concatenar los resultados al DataFrame original
df = pd.concat([df, resultados], axis=1)


df.head(5)
```
