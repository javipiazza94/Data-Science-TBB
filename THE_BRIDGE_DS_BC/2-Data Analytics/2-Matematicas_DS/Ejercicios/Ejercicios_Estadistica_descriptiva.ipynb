### Banco
La cantidad de dinero que se suele pedir en el banco Atope de Billets sigue la siguiente distribución. ¿Qué podemos sacar en claro de la distribución?

````python
datos_banco = np.random.normal(250000, 30000, 5000)
plt.hist(datos_banco, 70, histtype="stepfilled", alpha=.7);
````

Es un histograma que representa la dispersión de la cantidad de dinero que se suele pedir en el banco. Es una distribucion normal a la que le entran 2 parámetros: la medianda (250k), la desviación típica (30k) y el número de datos (5k). Es una distribucion más o menos simétrica, cuyos valores mínimos son cercanos a 150k y los máximos a 350k

### Cirujía
Suponiendo que el tiempo de supervivencia tras una operación quirúrgica de cerebro sigue la siguiente distribución. ¿Qué podemos sacar en claro?

````python
from scipy import stats
x = stats.gamma(5).rvs(5000)
plt.hist(x, 70, histtype="stepfilled", alpha=.7);
````

La mayoría de los supervivientes sobrevíve entre 2 y 6 años, siendo el intervalo entre 3 y 5 años el más habitual. A partir de los 6 años la curva baja progresivamente hasta los 14 años, siendo los 16 su máximo rango.

### Salarios
¿Qué podemos afirmar sobre la distribución de salarios, que quedaría representada de la siguiente manera?

````python
datos_banco = np.random.lognormal(10, 1, 5000)
plt.hist(datos_banco, 70, histtype="stepfilled", alpha=.7);
````

El tiesismo manda. Los salarios más habituales están en la franja entre 10 y 30k. A partir de los 40K la curva baja significativamente, siendo desde los 100k hasta los 350k su rango más alto, pero con muy poca concurrencia. 

## Ejercicio 2
Para este ejercicio no se pueden utilizar librerías ni funciones ya predefinidas. Solo las *built-in* de Python (sum, len, sort...). Math valdria si tienes que realizar alguna raiz cuadrada

Dados los siguientes datos de retrasos en vuelos de la compañia *Más vale tarde que nunca*:

```Python
retr = [3, 1, 1, 2, 4, 5, 6, 4, 7, 4, 5, 6, 4, 5, 6, 7, 4]
```

Se pide
* Calcular la media
* Calcular la moda
* Calcular la mediana
* Calcular el mayor retraso, el menor y el rango de la variable
* Calcula la varianza
* Calcular la desviación típica
* Prueba a calcular su histograma
* Interpreta los resultados

````python
retr = [3, 1, 1, 2, 4, 5, 6, 4, 7, 4, 5, 6, 4, 5, 6, 7, 4]
````

````python
#Media
numero_elementos = len(retr)
media = sum(retr)/numero_elementos
media
````

````python
#Moda

#Contamos elementos y los guardamos en un diccionario
frecuencias = {}
for valor in retr:
    frecuencias[valor] = retr.count(valor)
        
#Usamos la funcion max para encontrar el elemento que mas se repite       
moda = max(frecuencias, key=frecuencias.get)
resultado = f"La moda de la lista es {moda}"
moda
````

````python
#Mediana

# Calcula la mediana
if numero_elementos % 2 == 0:
    # Si la lista tiene un número par de elementos, la mediana es el promedio de los dos valores del medio
    mediana = (retr[numero_elementos // 2 - 1] + retr[numero_elementos // 2]) / 2
else:
    # Si la lista tiene un número impar de elementos, la mediana es el valor del medio
    mediana = retr[numero_elementos // 2]

mediana = f"Mediana de la lista es: {mediana}"
mediana
````

````python
#Usamos las funciones max, min y range
maximo = max(retr)
minimo = min(retr)
rango = range(minimo, maximo)
rango
````

````python
#Varianza de la lista
from math import pow
lista = retr

# Usamos suma para calcular la varianza
varianza = sum(pow((x - media), 2) for x in lista) / numero_elementos

resultado = f"Varianza de la lista es: {varianza}"
varianza
````

````python
#Desviacion tipica
from math import sqrt
Desviacion = sqrt(varianza)
Desviacion
````

````python
#Histograma

# Imprime el histograma
print("Histograma:")
for clave, valor in frecuencias.items():
    print(f"{clave}: {'*' * valor}")
````
