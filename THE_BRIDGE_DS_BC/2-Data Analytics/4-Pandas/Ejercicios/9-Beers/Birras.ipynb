## Ejercicio 14
Representa mediante un *boxplot* de `seaborn` la variable `abv`

```python
p_0 = np.percentile(nivel_alcohol, 0)  # Mínimo
p_25 = np.percentile(nivel_alcohol, 25)  # Q1
p_50 = np.percentile(nivel_alcohol, 50)  # Mediana (Q2)
p_75 = np.percentile(nivel_alcohol, 75)  # Q3
p_100 = np.percentile(nivel_alcohol, 100)  # Máximo

print("Valor del percentil 0 (Mínimo):", p_0)
print("Valor del percentil 25 (Q1):", p_25)
print("Valor del percentil 50 (Mediana):", p_50)
print("Valor del percentil 75 (Q3):", p_75)
print("Valor del percentil 100 (Máximo):", p_100)
print("Valor media:", round(cervezas['abv'].mean(), 2))
```

## Ejercicio 23
Consigue en un DataFrame todas las cervezas en cuyas cervecerías el mínimo nivel de `abv` sea de 0.07.

```python
min_abv_by_brewery = cervezas_datasets.groupby('brewery_name')['abv'].min()
```

## Ejercicio 24
Teniendo en cuenta la media de amargor de todas las cervezas, obtén una tabla con las cervezas cuyos estados tengan una media de amargor en cerveza por encima de la media

```python
media = cervezas_datasets['ibu'].mean()
```

```python
max_ibu = cervezas_datasets[cervezas_datasets['ibu'] > media]
tabla_media_amargor = max_ibu.groupby('state')['ibu'].mean().reset_index()

tabla_media_amargor
```
